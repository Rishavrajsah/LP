```python
# Generate list of transactions
transactions = []
for i in range(len(data)):
    transaction = [str(data.values[i,j]) for j in range(data.shape[1]) if pd.notna(data.values[i,j])]
    transactions.append(transaction)

# Display some sample transactions
print(transactions[:5])
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

# One-hot encoding of transactions
te = TransactionEncoder()
te_data = te.fit(transactions).transform(transactions)
df = pd.DataFrame(te_data, columns=te.columns_)

# Generate frequent itemsets using apriori
freq_itemsets = apriori(df, min_support=0.01, use_colnames=True)
print(freq_itemsets.head())
```

```python
# Manually set num_itemsets to the number of frequent itemsets
rules = association_rules(freq_itemsets, metric="confidence", min_threshold=0.2, num_itemsets=len(freq_itemsets))

# Display rules
print(rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']].head())
```

```python
import matplotlib.pyplot as plt

# Plotting Support vs Confidence
plt.scatter(rules['support'], rules['confidence'], alpha=0.5)
plt.xlabel('Support')
plt.ylabel('Confidence')
plt.title('Support vs Confidence')
plt.show()

# Plotting Lift vs Confidence
plt.scatter(rules['lift'], rules['confidence'], alpha=0.5, color='red')
plt.xlabel('Lift')
plt.ylabel('Confidence')
plt.title('Lift vs Confidence')
plt.show()
```

```python
# Generate rules with num_itemsets specified
high_confidence_rules = association_rules(freq_itemsets, metric="confidence", min_threshold=0.4, num_itemsets=len(freq_itemsets))
print(high_confidence_rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']].head())
```
